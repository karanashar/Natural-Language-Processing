```python
import nltk
from nltk.corpus import PlaintextCorpusReader
import os
mycorpus = PlaintextCorpusReader('.', '.*\.txt')
```

```python
mycorpus = mycorpus.raw('filename1.txt')
```

```python
token=nltk.sent_tokenize(mycorpus)
```

```python
import re
```

```python
str1=re.compile(('.+\?$'))
questions=[]
for i in token:
    if(re.match(str1,i)):
        questions.append(i)
```

```python
from nltk.corpus import treebank
a=treebank.tagged_sents()
t0 = nltk.DefaultTagger('NN')
t1 = nltk.UnigramTagger(a, backoff=t0)
t2 = nltk.BigramTagger(a, backoff=t1)
```

```python
questions_word=[nltk.word_tokenize(i) for i in questions]
```

```python
taggedtext = [t2.tag(token) for token in questions_word]
```

```python
grammar = r"""ADJ: {<JJ.*><NN.*>|<JJ.*><CC><JJ.*>?|<DT><JJ.*>}"""
cp = nltk.RegexpParser(grammar)
```

```python
question_adj=[]
for i in taggedtext:
    a=cp.parse(i)
    for subtree in a.subtrees(filter=lambda t : t.label()=='ADJ'):
        if(i in question_adj):
            continue
        else:
            question_adj.append(i)
        
```

```python
len(question_adj)
```

```python
sum1=0
for i in question_adj:
    a=len(i)
    sum1=sum1+a
average=sum1/len(question_adj)
print(average) 
```

```python
str2=re.compile(('.+!$'))
exclamation=[]
for i in token:
    if(re.match(str2,i)):
        exclamation.append(i)
```

```python
exclamation_word=[nltk.word_tokenize(i) for i in exclamation]
```

```python
taggedtext1 = [t2.tag(token) for token in exclamation_word]
```

```python
grammar1= r"""VB:{^<V.*>|^<DT><V.*>}"""
cp1=nltk.RegexpParser(grammar1)
imperative=[]
for i in taggedtext1:
    a=cp1.parse(i)
    for subtree in a.subtrees(filter=lambda t : t.label()=='VB'):
        imperative.append(i)    
```

```python
grammar2= r"""ADJ:{<JJ.*><NN.*>|<JJ.*><CC><JJ.*>?|<DT><JJ.*>}"""
cp2=nltk.RegexpParser(grammar2)
imperative_final=[]
for i in imperative:
    a=cp2.parse(i)
    for subtree in a.subtrees(filter=lambda t : t.label()=='ADJ'):
        if i in imperative_final:
            continue
        else:
            imperative_final.append(i) 
```

```python
len(imperative_final)
```

```python
sum1=0
for i in imperative_final:
    a=len(i)
    sum1=sum1+a
average=sum1/len(imperative_final)
print(average) 
```

<H1> Bigram Analysis

**Question Sentences**

```python
question_token=[]
for i in range(len(question_adj)):
    for j in range(len(question_adj[i])):
        question_token.append(question_adj[i][j][0])
```

```python
reviewbigrams = list(nltk.bigrams(question_token))
```

```python
reviewbigrams
```

```python
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(question_token)
```

```python
def Alpha(w):
    if(w.isalpha()):
        return False
    else:return True
```

```python
finder.apply_word_filter(Alpha)
finder.apply_word_filter(lambda w: w.lower() in stopwords)
scored = finder.score_ngrams(bigram_measures.raw_freq)
for bscore in scored[:50]:
    print (bscore)
```

```python
finder2 = BigramCollocationFinder.from_words(question_token)
finder2.apply_word_filter(Alpha)
finder2.apply_word_filter(lambda w: w.lower() in stopwords)
finder2.apply_freq_filter(5)
scored = finder.score_ngrams(bigram_measures.pmi)
for bscore in scored[:50]:
    print (bscore)
```
